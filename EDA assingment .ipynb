```python
#Which type of hotel is most preferred by the guests?
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='hotel')
plt.title("Hotel Type Preference")
plt.show()
```

```python
#What is percentage of hotel booking cancellation?
cancel = df['is_canceled'].value_counts(normalize=True)[1] * 100
print("Cancellation Rate: {cancel:.2f}%")
```

```python
#Which type of meal meal_preference = df['meal'].value_counts()
mp = df['meal'].value_counts()
mp = df['meal'].value_counts(normalize=True) * 100 
print("Meal Preference Count:\n", mp)
print("\nMeal Preference Percentage:\n", mp)
```

```python
# Count bookings by year
year = df['arrival_date_year'].value_counts().sort_index()
print("Bookings per Year:\n", year)
```

```python
#Which month has the most bookings in each hotel type?
# Group by hotel and month, then count bookings
monthly_bookings = df.groupby(['hotel', 'arrival_date_month']).size().reset_index(name='booking_count')

# To sort months in calendar order
import calendar
month_order = list(calendar.month_name)[1:]  # ['January', 'February', ..., 'December']
monthly_bookings['arrival_date_month'] = pd.Categorical(
    monthly_bookings['arrival_date_month'],
    categories=month_order,
    ordered=True
)

# Sort by hotel and month order
monthly_bookings = monthly_bookings.sort_values(['hotel', 'arrival_date_month'])

# Find the month with the max bookings for each hotel
most_booked_months = monthly_bookings.loc[monthly_bookings.groupby('hotel')['booking_count'].idxmax()]

print("Month with Most Bookings for Each Hotel Type:\n", most_booked_months)
```

```python
#From which country most guests come?'

tc = df['country'].value_counts().head(10)
print("Top 10 Countries by Number of Guests:", tc)
```

```python
#Which distribution channel is most used in booking?
cu = df['distribution_channel'].value_counts()
cp= df['distribution_channel'].value_counts(normalize=True) * 100

print("Distribution Channel Usage:\n", cu)
print("\nDistribution Channel Percentage:\n", cp)
```

```python
#Which room type is most preferred by guests?
# Most preferred room types (booked by guests)
room_p = df['reserved_room_type'].value_counts()
prefer_room= df['reserved_room_type'].value_counts(normalize=True) * 100

print("Preferred Room Types (Count):\n", room_p)
print("\nPreferred Room Types (Percentage):\n", prefer_room)
```

```python
#Which room type is most assigned?
assigned_room_types = df['assigned_room_type'].value_counts()
assigned_room_types_percentage = df['assigned_room_type'].value_counts(normalize=True) * 100

print("Assigned Room Types (Count):\n", assigned_room_types)
print("\nAssigned Room Types (Percentage):\n", assigned_room_types_percentage)
```

```python
#Top 5 agents in terms of most bookings?
agents = df['agent'].value_counts().head(5)
print("Top 5 Agents by Number of Bookings:\n", agents)
```

```python
#What is the percentage of repeated guests?
guest1 = df['is_repeated_guest'].value_counts()
guest2 = df['is_repeated_guest'].value_counts(normalize=True) * 100
print("Repeated Guest Count:\n", guest1)
print("\nRepeated Guest Percentage:\n", guest2)
```

```python
#12. Which customer type has the most booking?
customer1 = df['customer_type'].value_counts()
customer2 = df['customer_type'].value_counts(normalize=True) * 100

print("Customer Type Counts:\n", customer1)
print("\nCustomer Type Percentages:\n", customer2)
```

```python
#Which Market Segment has the most booking?

n = df['market_segment'].value_counts()
m = df['market_segment'].value_counts(normalize=True) * 100

print("Market Segment Counts:\n", n)
print("\nMarket Segment Percentages:\n", m)
```

```python
# Most common deposit type
deposit1 = df['deposit_type'].value_counts()
deposit2 = df['deposit_type'].value_counts(normalize=True) * 100

print("Deposit Type Counts:\n", deposit1)
print("\nDeposit Type Percentages:\n", deposit2)
```

```python
# Total stay duration (weekend + week nights)
df['total_stay'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']

# Average stay
average_stay = df['total_stay'].mean()

print(f"Average Stay Duration: {average_stay:.2f} nights")
```

```python
# Revenue = adr * total_stay
df['revenue'] = df['adr'] * df['total_stay']

# Total revenue by hotel
revenue_by_hotel = df.groupby('hotel')['revenue'].sum().sort_values(ascending=False)

print("Total Revenue by Hotel:\n", revenue_by_hotel)
```

```python
# Revenue = adr * total_stay
df['revenue'] = df['adr'] * df['total_stay']

# Total revenue by hotel
revenue_by_hotel = df.groupby('hotel')['revenue'].sum().sort_values(ascending=False)

print("Total Revenue by Hotel:\n", revenue_by_hotel)
```

```python
# Average waiting list time by hotel
waiting_by_hotel = df.groupby('hotel')['days_in_waiting_list'].mean()

print("Average Waiting Days by Hotel:\n", waiting_by_hotel)
```

```python
# Count repeated guests by hotel
repeated_by_hotel = df[df['is_repeated_guest'] == 1].groupby('hotel')['is_repeated_guest'].count()
print("Repeated Guest Count by Hotel:\n", repeated_by_hotel)
```

```python
# Average ADR per month (across all years)
monthly_adr = df.groupby('arrival_date_month')['adr'].mean()

# Ensure calendar order
import calendar
month_order = list(calendar.month_name)[1:]
monthly_adr = monthly_adr.reindex(month_order)

print("Average ADR by Month:\n", monthly_adr)
```

```python
# Average ADR by distribution channel
adr_by_channel = df.groupby('distribution_channel')['adr'].mean().sort_values(ascending=False)

print("ADR by Distribution Channel:\n", adr_by_channel)
```
